# 卷积神经网络

- 卷积神经网络(CNN: Convolutional Neural Networks)是计算机视觉领域最经典的模型结构，包含卷积、池化、激活函数、批归一化、丢弃法等常用模块。
- 使用卷积神经网络进行特征提取，既能提取到相邻像素点之间到特征模式，又能保证参数的个数不随图片尺寸变化。
- 典型的卷积神经网络结构：多层卷积和池化层组合作用在输入图片上，在网络的最后通常会加入一系列全连接层，ReLU激活函数一般加在卷积或者全连结层的输出上，网络中通常还会加入Dropout来防止过拟合。



## 卷积计算
- 卷积是数学分析中的一种积分变换的方法，在图像处理中采用的是卷积的离散形式。
- 在卷积神经网络中，卷积层的实现方式实际上是数学中定义的互相关 （cross-correlation）运算，与数学分析中的卷积定义有所不同。

- 卷积核（kernel）也被叫做滤波器（filter），假设卷积核的高和宽分别为h和w，则称为 h * w 卷积。 
- 在卷积神经网络中，一个卷积算子除了上面描述的卷积过程之外，还包括加上偏执项的操作。

## 填充 Padding

- 经过一次卷积，图片尺寸变小，卷积输出特征图的尺寸计算方法：输出图w/h = 输入图w/h - 卷积核w/h + 1 (步幅为1时的情况)。
- 经过多层卷积，输出图片尺寸会不断减小，通常会在图片的外围进行填充，以保持图像尺寸不变。

## 批量处理多输入/多输出通道



# 池化 Pooling

- 池化是使用某一位置的相邻输出的总体统计特征代替网络在该位置的输出，其好处是当输入数据做出少量平移时，经过池化函数后的大多数输出还能保持不变。
- 池化之后的特征图会变小，如果后面连接的是全连接层，能有效减少神经元的个数，节省存储空间并提高计算效率。
- 池化窗口在图片上滑动时，与卷积核类似，也有步幅和填充的概念。
- 常用的池化方法有平均池化和最大池化。



# 批归一化 Batch Normalization

- 批归一化方法已被广泛应用在深度学习中，其目的是对神经网络中间层的输出进行标准化处理，使得中间层的输出更加稳定。
- 通常会对神经网络的数据进行标准化处理，处理后的样本数据集满足均值为0，方差为1的统计分布，这是因为当输入数据的分布比较固定时，有利于算法的稳定和收敛。
- 对于深度神经网络来说，由于参数是不断更新的，即使输入数据已经做过标准化处理，但是对于比较靠后的那些层，其接收到的输入仍然是剧烈变化的，通常会导致数值不稳定，模型很难收敛。
- 批归一化能够使神经网络中间层的输出变得更加稳定，并有如下三个优点：
    - 使学习快速进行（能够使用较大的学习率）
    - 降低模型对初始值的敏感性
    - 在一定程度上抑制过拟合
- 对于深度神经网络来说，由于参数是不断更新的，即使输入数据已经做过标准化处理，但是对于比较靠后的那些层，其接收到的输入仍然是剧烈变化的，通常会导致数值不稳定，模型很难收敛。

# 丢弃法 Dropout

- 丢弃法（Dropout）是深度学习中一种常用的抑制过拟合的方法，其做法是在神经网络学习过程中，随机删除一部分神经元。
- 训练时，随机选出一部分神经元，将其输出设置为0，这些神经元将不对外传递信号。
- 丢弃法通过随机将输入数据的元素置0，消除减弱了神经元节点间的联合适应性，增强模型的泛化能力。
