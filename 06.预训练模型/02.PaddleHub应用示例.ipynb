## 加载PaddleHub

```python
import paddlehub as hub

# 检查本地与远端PaddleHub-Server的连接状态
paddlehub.server_check()
```

## 计算机视觉

```python
img_path = './data/test.jpg'

# 原图展示
import matplotlib.image as mpimg
import matplotlib.pyplot as plt

img = mpimg.imread(img_path)

plt.figure(figsize=(8, 8))
plt.imshow(img)
plt.axis('off')
plt.show()
```

### 人脸检测

```python
# 人脸检测
module = hub.Module(name='ultra_light_fast_generic_face_detector_1mb_640')
res = module.face_detection(paths=['./data/test.jpg'], visualization=True, output_dir='output/face')

res_img_path = './output/face/test.jpg'
img = mpimg.imread(res_img_path)

plt.figure(figsize=(8, 8))
plt.imshow(img)
plt.axis('off')
plt.show()
```

### 关键点检测

```python
module = hub.Module(name='human_pose_estimation_resnet50_mpii')
res = module.keypoint_detection(paths=['./data/test.jpg'], visualization=True, output_dir='output/keypoint')

res_img_path = './output/keypoint/test.jpg'
img = mpimg.imread(res_img_path)

plt.figure(figsize=(8, 8))
plt.imshow(img)
plt.axis('off')
plt.show()
```

### 人体部位分隔

```python
module = hub.Module(name='ace2p')
res = module.segmentation(paths=['./data/test.jpg'], visualization=True, output_dir='output/ace2p')

res_img_path = './output/ace2p/test.png'
img = mpimg.imread(res_img_path)

plt.figure(figsize=(8, 8))
plt.imshow(img)
plt.axis('off')
plt.show()
```

### 人像抠图

```python
module = hub.Module(name='deeplabv3p_xception65_humanseg')
res = module.segmentation(paths=['./data/test.jpg'], visualization=True, output_dir='output/humanseg')

res_img_path = './output/humanseg/test.png'
img = mpimg.imread(res_img_path)

plt.figure(figsize=(8, 8))
plt.imshow(img)
plt.axis('off')
plt.show()
```

## 自然语言处理 NLP

### 中文分词

```python
lac = hub.Module(name='lac')

test_text = ['对于自然语言处理任务，下面以中文分词和情感分类的任务为例，待处理的数据以函数参数的形式传入。']

res = lac.lexical_analysis(texts=test_text)

print('中文分词结果：', res)
```

### 情感分类

```python
senta = hub.Module(name='senta_bilstm')

test_text = ['马斯克表示对OpenAI的现状感到失望，他认为OpenAI已经背离了其开源和非营利的理念，变成了一个由微软控制的利润最大化的公司。']

res = senta.sentiment_classify(texts=test_text)

print('情感分类结果：', res)
```
