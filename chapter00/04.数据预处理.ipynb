## 图像数据预处理

- 解决过拟合问题的方式之一就是对训练数据做增强处理，以增加样本的多样性，从而增强模型的泛化能力。
- 飞浆内置了数十种图像数据处理方法，包括随机裁剪、旋转变换、图像亮度、图像对比度等。

```python
import paddle
```

```python
print('图像数据处理方法：', paddle.vision.transforms.__all__)
```

```python
# 单个使用
from paddle.vision.transforms import Resize

transform = Resize(size=32)
```

```python
# 多个使用
from paddle.vision.transforms import Compose, RandomRotation

transform = Compose([RandomRotation(10), Resize(size=32)])
```

## 图像预处理示例

```python
from paddle.vision.transforms import CenterCrop
import matplotlib.pyplot as plt
import cv2
```

```python
image = cv2.imread('../datasets/flower.png')  # cv2.IMREAD_GRAYSCALE

image.shape
```

```python
# 按指定尺寸对图像进行裁剪，保持图像中心点不变
transform = CenterCrop(200)

image2 = transform(image)

image2.shape
```

```python
plt.subplot(1, 2, 1)
plt.title('Origin Image')
plt.imshow(image[:, :, ::-1])

plt.subplot(1, 2, 2)
plt.title('CenterCrop Image')
plt.imshow(image2[:, :, ::-1])
```
