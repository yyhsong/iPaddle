```python
from paddle.vision.transforms import CenterCrop
import matplotlib.pyplot as plt
import cv2
```

```python
image = cv2.imread('../datasets/flower.png')  # cv2.IMREAD_GRAYSCALE

image.shape
```

```python
# 
transform = CenterCrop(200)

image2 = transform(image)

image2.shape
```

```python
plt.subplot(1, 2, 1)
plt.title('Origin Image')
plt.imshow(image[:, :, ::-1])

plt.subplot(1, 2, 2)
plt.title('CenterCrop Image')
plt.imshow(image2[:, :, ::-1])
```
